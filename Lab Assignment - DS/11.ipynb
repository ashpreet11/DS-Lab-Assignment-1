Singly Linked List without Class using Dictionary

```python
n1 = {"data":10, "next":None}
n2 = {"data":20, "next":None}
n3 = {"data":30, "next":None}
```

```python
n2["next"] = n3
n1["next"] = n2 

head = n1
```

```python
temp = head
while temp is not None: 
    print(temp["data"])
    temp = temp["next"] 
```

```python
# Program to Demonstrate Basic Operations on Data Structures (List) 

l = []

print("Original List:", l)

# Add element
l.append(60)
l.append(90)
l.append(30)
l.append(10)
print("After append:", l)

# Insert element
l.insert(2, 25)
print("After insert at index 2:", l)

# Remove element (by value)
l.remove(30)
print("After removing 30:", l)

# Delete element (by index)
l.pop(1)
print("After deleting index 1:", l)

# Search element
if 40 in l:
    print("40 found at index:", l.index(40))
else:
    print("40 not found")

# Update element
l[0] = 5
print("After updating index 0:", l)

# Length of list
print("Length of list:", len(l))

# Sort list
l.sort()
print("After sorting:", l)

# Reverse list
l.reverse()
print("After reversing:", l)
```

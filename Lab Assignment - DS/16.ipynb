```python
# Insertion Sort Implementation in Python

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

    return arr

arr = [64, 25, 12, 22, 11,50,24,60,17,30]
print("Sorted array:", insertion_sort(arr))
```

```python
def binary_search(arr, target):
    low = 0 
    high = len(arr) - 1
    
    while low <= high:
        mid = low + (high - low) // 2
        
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
            
    return -1
arr = [10, 20, 30, 40, 50]
print(binary_search(arr, 40))
```

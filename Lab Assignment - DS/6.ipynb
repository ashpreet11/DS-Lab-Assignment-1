Postfix Expression Evaluation using Stack

```python
def evaluate_postfix(expression):
    stack = []

    for char in expression:    
        if char.isdigit():
            stack.append(int(char))
        else:
            b = stack.pop()
            a = stack.pop()

            if char == '+':
                result = a + b    
            elif char == '-':
                result = a - b    
            elif char == '*':
                result = a * b    
            elif char == '/':
                result = a / b

    return stack.pop()
    
expression = "5 6  2 + * 12 4 / -"
print(evaluate_postfix(expression))                     
```

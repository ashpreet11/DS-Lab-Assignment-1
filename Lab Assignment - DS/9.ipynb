Implement row-major using python code

1. Traversing a 2D list in row-major order

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

rows = len(matrix)
cols = len(matrix[0])

for i in range(rows):          # row loop
    for j in range(cols):      # column loop
        print(matrix[i][j], end=" ")
```

2. Converting 2D indices → 1D (row-major mapping)

```python
def to_1d(i, j, cols):
    return i * cols + j

print(to_1d(2, 1, 3))  # Output: 7
```

3. Converting 1D index → 2D (row-major mapping)

```python
def to_2d(index, cols):
    i = index // cols
    j = index % cols
    return i, j

print(to_2d(7, 3))  # Output: (2, 1)
```

4. Flattening a matrix in row-major order

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6]
]

flat = [element for row in matrix for element in row]
print(flat)  # [1, 2, 3, 4, 5, 6]
```

5. Creating a matrix from a 1D list (row-major fill)

```python
def create_matrix(arr, rows, cols):
    matrix = []
    for i in range(rows):
        row = arr[i * cols:(i + 1) * cols]
        matrix.append(row)
    return matrix

arr = [1, 2, 3, 4, 5, 6]
print(create_matrix(arr, 2, 3))
```

Enqueue and Dequeue

```python
# Queue implementation using a list

queue = []

# Enqueue operation
def enqueue(item):
    queue.append(item)
    print(f"{item} added to queue")

# Dequeue operation
def dequeue():
    if len(queue) == 0:
        print("Queue is empty")
    else:
        item = queue.pop(0)
        print(f"{item} removed from queue")

# Example usage
enqueue(10)
enqueue(20)
enqueue(30)

print("Queue:", queue)

dequeue()
print("Queue after dequeue:", queue)
```

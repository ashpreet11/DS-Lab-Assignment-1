Program to Implement Binary Search Tree (BST)

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

def insert(root, key):
    if root is None:
        return Node(key)

    if key < root.key:
        root.left = insert(root.left, key)
    else:
        root.right = insert(root.right, key)

    return root

def inorder(root):
    if root:
        inorder(root.left)
        print(root.key, end=" ")
        inorder(root.right)

# Driver Code
root = None
values = [50, 30, 70, 20, 40, 60, 80]

for value in values:
    root = insert(root, value)

print("BST Inorder Traversal:")
inorder(root)
```
